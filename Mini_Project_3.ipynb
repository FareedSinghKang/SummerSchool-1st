```python
from itertools import combinations

S = {-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}

result = []

for size in range(3, 7):  # Sizes 3 to 6 inclusive
    for subset in combinations(S, size):
        if sum(subset) == 0:
            result.append(subset)

for r in result:
    print(r)
```
